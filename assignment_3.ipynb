2. Line chart: A line chart is a visualization that shows the relationship between two variables over time. To create a line chart in Python, you can use the plot.line() method in Pandas.

```python
# Create a line chart of the `salary` column over time
df1['Salary'].plot.line()

# Add a title and labels to the chart
plt.title('Salary Over Time')
plt.xlabel('Year')
plt.ylabel('Salary')

# Show the chart
plt.show()
```

3. Pie chart: A pie chart is a visualization that shows the relative size of different categories. To create a pie chart in Python, you can use the plot.pie() method in Pandas.

```python
# Create a pie chart of the `salary` column
df1['GenderID'].plot.pie()

# Add a title and labels to the chart
plt.title('GenderID')
plt.legend()

# Show the chart
plt.show()
```
