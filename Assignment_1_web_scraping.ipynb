```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
# URL to scrape
url = 'http://quotes.toscrape.com/page/1/'

```

```python
# Send a GET request to the URL and save the response to a variable
response = requests.get(url)
```

```python
response.content
```

```python
# Create a BeautifulSoup object from the HTML content of the response
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
# Find all the div elements with a class attribute of "quote"
quotes = soup.find_all('div', class_='quote')
```

```python
data = []

for quote in quotes:
    author = quote.find('small', class_='author').text
    text = quote.find('span', class_='text').text
    data.append({'author': author, 'text': text})
```

```python
df = pd.DataFrame(data)

print(df.head())
```

```python
#another method 
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://quotes.toscrape.com/'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
quotes = soup.find_all('div', class_='quote')
```

```python
data = []
```

```python
for quote in quotes:
    quote_data = {
        'quote': quote.find('span', class_='text').text,
        'author': quote.find('small', class_='author').text,
        'tags': quote.find_all('a', class_='tag')
    }
    data.append(quote_data)
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv('quotes.csv', index=False)
```
